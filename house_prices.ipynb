## 3.2 模型
### 3.2.1 机器学习 sklearn

```python
# 导入学习库
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold
from sklearn.metrics import make_scorer

# 设置kfold，交叉采样
kfold = KFold(n_splits=10, shuffle=True, random_state=1)

# 设置回归器
regressors = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    ElasticNet(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
    KNeighborsRegressor(),
    DecisionTreeRegressor(),
    SVR(),
    ExtraTreesRegressor(),
]
```

```python
def log_rmse(y_true, y_pred):
    # 将小于1的预测值设为1，以避免取对数时出现负无穷大
    clipped_preds = np.maximum(y_pred, 1)
    log_rmse_value = np.sqrt(np.mean((np.log(clipped_preds) - np.log(y_true)) ** 2))
    return log_rmse_value

# 创建一个自定义评分器
log_rmse_scorer = make_scorer(log_rmse, greater_is_better=False)
```

```python
#交叉验证结果
cv_results = {}
for regressor in regressors:
    name = regressor.__class__.__name__
    cv_results[name] = cross_val_score(regressor,train_dummy_X,train_logy,scoring=log_rmse_scorer,cv=kfold,n_jobs=-1)

# print(cv_results)

# 打印结果
for name, scores in cv_results.items():
    print(f"{name}: Mean RMSLE = {-scores.mean():.4f}, Std RMSLE = {scores.std():.4f}")
```

```python
# 训练  使用GradientBoostingRegressor
# 设置参数
params_GradientBoostingRegressor = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1],
    'max_features': [0.8, 0.9, 1]
}
# 设置模型
model_GradientBoostingRegressor = GradientBoostingRegressor()
#网格搜索
#参数分别是模型，参数列表，交叉验证，评分，设置为-1表示使用所有的CPU进行训练，verbose表示输出日志
grid_GradientBoostingRegressor = GridSearchCV(model_GradientBoostingRegressor, params_GradientBoostingRegressor, scoring=log_rmse_scorer, cv=kfold, n_jobs=-1,verbose=1)
# 训练
grid_GradientBoostingRegressor.fit(train_dummy_X, train_logy)
#输出结果
grid_GradientBoostingRegressor.best_score_
```

```python
# 输出最佳参数
print(grid_GradientBoostingRegressor.best_params_)
```

```python
# 使用最佳参数训练模型
model_GradientBoostingRegressor = GradientBoostingRegressor(**grid_GradientBoostingRegressor.best_params_)
model_GradientBoostingRegressor.fit(train_dummy_X, train_logy)
```

```python
# 预测
test_dummy_X = full_dummy.loc[test.index]
```

```python
preds = model_GradientBoostingRegressor.predict(test_dummy_X)
preds = np.expm1(preds)
```

```python
# 保存结果
submission = pd.DataFrame({'Id': test.index, 'SalePrice': preds})
submission.to_csv('submission.csv', index=False)
submission.describe()
```

### 3.3.2 深度学习 pytorch

直接使用一个基本的线性回归模型和平方损失函数来训练模型

```python
def log_rmse(net, features, labels):
    with torch.no_grad():
        # 将小于1的值设成1，使得取对数时数值更稳定
        clipped_preds = torch.max(net(features), torch.tensor(1.0))
        rmse = torch.sqrt(loss(clipped_preds.log(), labels.log()))
    return rmse.item()
```

```python
import torch 
import torch
import torch.nn as nn
```

```python
loss = torch.nn.MSELoss()
```

```python
net=nn.Sequential(nn.Linear(train_dummy_X.shape[1],1))
net
```

```python
optimizer = torch.optim.Adam(net.parameters(),lr=0.01)
```

```python
num_epochs = 300
```

```python
net.to('cpu')
def train(train_iter,test_iter,net,loss,optimizer,num_epochs):
    # net.to('cuda')
    # train_l = 0
    for epoch in range(num_epochs):
        for X,y in train_iter:
            # X = X.to('cuda')
            # y = y.to('cuda')
            l = loss(net(X),y)
            optimizer.zero_grad()
            l.backward()
            optimizer.step()
            # train_l += log_rmse(net,X,y)
        train_l = log_rmse(net,train_iter.dataset.tensors[0],train_iter.dataset.tensors[1])
        if test_iter is not None:
            test_l = log_rmse(net,test_iter.dataset.tensors[0],test_iter.dataset.tensors[1])
            if (epoch+1)%100==0:
                print(f'epoch {epoch+1},train log rmse {train_l:.6f},test log rmse {test_l:.6f}')
        else:
            if (epoch+1)%100==0:
                print(f'epoch {epoch+1},train log rmse {train_l:.6f}')
```

```python
def get_K_fold_data(k,i,X,y):
    n = X.shape[0]
    fold_size = n//k
    id = fold_size*i
    test_X = X[id:id+fold_size]
    test_y = y[id:id+fold_size]
    train_X = torch.cat((X[:id],X[id+fold_size:]),dim=0)
    train_y = torch.cat((y[:id],y[id+fold_size:]),dim=0)
    return train_X,train_y,test_X,test_y
```

```python
train_dummy_X = train_dummy_X.astype(float)
X = torch.tensor(train_dummy_X.values,dtype=torch.float32)

train_logy = train_logy.astype(float)
y = torch.tensor(train_logy.values,dtype=torch.float32).view(-1,1)
```

```python
def K_fold(k,batch_size,X,y):
    for i in range(k):
        print(f'第{i+1}折交叉验证')
        train_X,train_y,test_X,test_y = get_K_fold_data(k,i,X,y)
        train_dataset = torch.utils.data.TensorDataset(train_X,train_y)
        test_dataset = torch.utils.data.TensorDataset(test_X,test_y)
        train_iter = torch.utils.data.DataLoader(train_dataset,batch_size,shuffle=True)
        test_iter = torch.utils.data.DataLoader(test_dataset,batch_size,shuffle=False)
        train(train_iter,test_iter,net,loss,optimizer,num_epochs)
```

```python
K_fold(5,256,X,y)
```

```python
# 预测
test_dummy_X = full_dummy.loc[test.index]
test_dummy_X = test_dummy_X.astype(float)
test_X = torch.tensor(test_dummy_X.values,dtype=torch.float32)
preds = net(test_X).detach().numpy()
preds = np.expm1(preds)

preds
```

```python
submission = pd.DataFrame({'Id': test.index, 'SalePrice': preds[:,0]})
submission.to_csv('submission.csv', index=False)
```
